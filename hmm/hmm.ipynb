```python
def selectDistribution():
    return 1 if numpy.random.random()<.5 else 0

l_state= [3, 8]
l_len= [15, 7]


seq= []
inds= []
for i in range(100):
    state= selectDistribution()
    len_sel= numpy.random.poisson(l_len[state])
    inds.extend([state]*len_sel)
    seq.extend(numpy.random.poisson(l_state[state], len_sel))
```

```python
plot(seq)
```

```python
def logSumExp(logs):
    '''
        Arguments:
            logs - a list of log(probabilities)
        Returns: 
            log(sum(exp(logs)))
    '''
    a= min(logs)
    return a+log(sum(exp(i-a) for i in logs))
```
